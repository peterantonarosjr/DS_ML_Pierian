# Cross Validation & Bias Variance Tradeoff

## Bias Variance Tradeoff
* The bias-variance trade-off is the pooint where we are adding noise because of the increase in model complexity
* The training error will continuously go down as we increase model complexity, but our test error will begin to increase
* Underfitting -> Overfitting
* If we imagine repeating the modeling process each time being an individual realization of the training data, we can begin to see the points scattered around the optimal error.

## Cross Validation
* Cross Validation reduces variance as it uses separate splits for both the training and test data
* These separate splits ensure we don't get "lucky" or "unlucky" on our training and test sets
* Overall our error seen at the end of cross validation will most likely be worse than our best error, but better than our worse error
* Cross validated models are better error aproximations to how the model will behave in the real world
